```python
def analizuj_tekst(tekst_wejsciowy):
    # KROK 1: Liczymy słowa
    # Metoda .split() dzieli tekst tam, gdzie są spacje
    lista_slow = tekst_wejsciowy.split()
    liczba_slow = len(lista_slow)

    # KROK 2: Liczymy zdania
    # Instrukcja mówi: użyj .split(".") - czyli tniemy tekst tam, gdzie są kropki
    lista_zdan = tekst_wejsciowy.split(".")

    # Uwaga: split po kropce często zostawia na końcu pusty element.
    # Żeby wynik był dokładny, odejmujemy 1 od długości listy (jeśli tekst kończy się kropką)
    # LUB po prostu przyjmujemy długość tej listy minus 1.
    # Dla uproszczenia tutaj przyjmiemy prostą metodę:
    liczba_zdan = len(lista_zdan) - 1

    # Jeśli wyjdzie 0 (np. brak kropek), poprawiamy na 0 (zabezpieczenie)
    if liczba_zdan < 0:
        liczba_zdan = 0

    # KROK 3: Zwracamy wynik jako słownik (dict)
    return {"slowa": liczba_slow, "zdania": liczba_zdan}

# --- CZĘŚĆ GŁÓWNA PROGRAMU ---

# 1. Definiujemy przykładowy tekst (możesz tu wpisać cokolwiek)
moj_tekst = "Uczę się języka Python. To jest zadanie finałowe. Idzie mi całkiem nieźle."

# 2. Wywołujemy funkcję i zapisujemy wynik do zmiennej
wynik_analizy = analizuj_tekst(moj_tekst)

# 3. Wypisujemy raport wyciągając dane ze słownika
print(f"Tekst zawiera {wynik_analizy['slowa']} słów i {wynik_analizy['zdania']} zdań.")
```
